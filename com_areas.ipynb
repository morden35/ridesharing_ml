```python
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
cmap = pd.read_csv('data/community_area/2014_2018.csv')
cmap.columns = cmap.columns.str.lower()
cmap["geog"] = cmap["geog"].str.lower()

cmap["geog"] = cmap["geog"].str.replace("'", '')
cmap["geog"] = np.where(cmap.geog == "the loop", "loop", cmap.geog)

cmap.head(5)
```

```python
cmap.info()
```

```python
com_boundaries = gpd.read_file("data/community_area/com_boundaries.geojson")
com_boundaries["community"] = com_boundaries["community"].str.lower()
com_boundaries = com_boundaries.drop(['area', 'shape_area', 'perimeter', 
                                      'area_num_1', 'comarea_id', 'comarea', 
                                      'shape_len'], axis=1)
com_boundaries.columns = ['geog', 'area_num', 'geometry']
com_names = set(com_boundaries.geog)
cmap_names = set(cmap.geog)

# Checking for naming differences
print(cmap_names.difference(com_names))
```

```python
# Creating new features

cmap["perc_white"] = (cmap["white"] / cmap["tot_pop"]) * 100
cmap["perc_black"] = (cmap["black"] / cmap["tot_pop"]) * 100
cmap["perc_asian"] = (cmap["asian"] / cmap["tot_pop"]) * 100
cmap["perc_hisp"] = (cmap["hisp"] / cmap["tot_pop"]) * 100
cmap["perc_other_race"] = (cmap["other"] / cmap["tot_pop"]) * 100
cmap["perc_und19"] = (cmap["und19"] / cmap["tot_pop"]) * 100
cmap["perc_a20_34"] = (cmap["a20_34"] / cmap["tot_pop"]) * 100
cmap["perc_a35_49"] = (cmap["a35_49"] / cmap["tot_pop"]) * 100
cmap["perc_a50_64"] = (cmap["a50_64"] / cmap["tot_pop"]) * 100
cmap["perc_a65_74"] = (cmap["a65_74"] / cmap["tot_pop"]) * 100
cmap["perc_a75_84"] = (cmap["a75_84"] / cmap["tot_pop"]) * 100
cmap["perc_ov85"] = (cmap["ov85"] / cmap["tot_pop"]) * 100
cmap["perc_in_lbfrc"] = (cmap["in_lbfrc"] / cmap["tot_pop"]) * 100
cmap["perc_emp"] = (cmap["emp"] / cmap["tot_pop"]) * 100
cmap["perc_unemp"] = (cmap["unemp"] / cmap["tot_pop"]) * 100
cmap["perc_not_in_lbfrc"] = (cmap["not_in_lbfrc"] / cmap["tot_pop"]) * 100
cmap["perc_work_at_home"] = (cmap["work_at_home"] / cmap["tot_pop"]) * 100
cmap["perc_tot_comm"] = (cmap["tot_comm"] / cmap["tot_pop"]) * 100
cmap["perc_drove_al"] = (cmap["drove_al"] / cmap["tot_pop"]) * 100
cmap["perc_carpool"] = (cmap["carpool"] / cmap["tot_pop"]) * 100
cmap["perc_transit"] = (cmap["transit"] / cmap["tot_pop"]) * 100
cmap["perc_walk_bike"] = (cmap["walk_bike"] / cmap["tot_pop"]) * 100
cmap["perc_comm_other"] = (cmap["comm_other"] / cmap["tot_pop"]) * 100
cmap["agg_tt_avg"] = (cmap["agg_tt"] / cmap["emp"]) * 100
cmap["perc_lt_hs"] = (cmap["lt_hs"] / cmap["pop_25ov"]) * 100
cmap["perc_hs"] = (cmap["hs"] / cmap["pop_25ov"]) * 100
cmap["perc_some_coll"] = (cmap["some_coll"] / cmap["pop_25ov"]) * 100
cmap["perc_assoc"] = (cmap["assoc"] / cmap["pop_25ov"]) * 100
cmap["perc_bach"] = (cmap["bach"] / cmap["pop_25ov"]) * 100
cmap["perc_grad_prof"] = (cmap["grad_prof"] / cmap["pop_25ov"]) * 100
cmap["perc_ct_1phh"] = (cmap["ct_1phh"] / (cmap["ct_1phh"]+cmap["ct_2phh"]+cmap["ct_3phh"]+cmap["ct_4mphh"])) * 100
cmap["perc_ct_2phh"] = (cmap["ct_2phh"] / (cmap["ct_1phh"]+cmap["ct_2phh"]+cmap["ct_3phh"]+cmap["ct_4mphh"])) * 100
cmap["perc_ct_3phh"] = (cmap["ct_3phh"] / (cmap["ct_1phh"]+cmap["ct_2phh"]+cmap["ct_3phh"]+cmap["ct_4mphh"])) * 100
cmap["perc_ct_4mphh"] = (cmap["ct_4mphh"] /(cmap["ct_1phh"]+cmap["ct_2phh"]+cmap["ct_3phh"]+cmap["ct_4mphh"])) * 100
```

```python
to_drop = ['tot_pop', 'und19', 'a20_34', 'a35_49', 'a50_64', 'a65_74',
           'a75_84', 'ov85', 'pop_25ov', 'white', 'hisp', 'black',
           'asian', 'other', 'in_lbfrc', 'emp', 'unemp', 'not_in_lbfrc',
           'work_at_home', 'tot_comm', 'drove_al', 'carpool', 'transit',
           'walk_bike', 'comm_other', 'no_veh', 'one_veh', 'two_veh',
           'threeom_veh', 'lt_hs', 'hs', 'some_coll', 'assoc', 'bach', 'grad_prof',
           'inc_lt_25k', 'inc_25_50k', 'inc_50_75k', 'inc_75_100k', 'inc_100_150k',
           'inc_gt_150', 'ct_1phh', 'ct_2phh', 'ct_3phh', 'ct_4mphh', 'hu_tot', 'agg_tt']
# Remove two columns name is 'C' and 'D'
cmap = cmap.drop(to_drop, axis = 1)
```
